```python
# Import libraries & module dependencies
import pandas as pd
import glob
import os
```

```python
# get data file names
path =r'Data'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))
```

```python
# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
```

```python
print(len(big_frame))
```

```python
big_frame.drop_duplicates("Date", inplace=True)
#big_frame.set_index("Date", inplace=True)

big_frame.info()
```

```python
big_frame.head()
```

```python
big_frame.rename(columns={'Trips over the past 24-hours (midnight to 11:59pm)':'Trips',
                     'Miles traveled today (midnight to 11:59 pm)':'Distance(mi)',
                     'Total Annual Members (All Time)':'Membership',
                     '24-Hour Passes Purchased (midnight to 11:59 pm)':'1-Day Pass',
                     '3-Day Passes Purchased (midnight to 11:59 pm)':'3-Day Pass'}, inplace=True)
```
